```python
import networkx as nx
F = nx.Graph()

F.add_nodes_from([1,2,3,4])

F.add_edges_from([(1,2),(2,3),(1,3),(1,4)])

nx.draw(F, with_labels=True,node_color='#7209b7',
        node_size=1600,
        font_color='white',
        font_size=16)
```

```python
nx.has_path(F, 3, 4)
```

```python
list(nx.all_simple_paths(F, 3, 4))
```

```python
nx.shortest_path(F, 3, 4)
```

```python
nx.shortest_path_length(F, 3, 4)
```

```python
nx.is_connected(F)
```

```python
F = nx.Graph()
nx.add_cycle(F,(1,2,3))
F.add_edge(4,5)
nx.draw(F, with_labels=True,node_color='#7209b7',
        node_size=1600)
```

```python
nx.is_connected(F)
```

```python
nx.has_path(F, 3, 5)
```

```python
nx.number_connected_components(F)
```

```python
list(nx.connected_components(F))
```

```python
components = list(nx.connected_components(F))
len(components[0])
```

```python
max(nx.connected_components(F), key=len)
```

```python
core_nodes = max(nx.connected_components(F), key=len)
core = F.subgraph(core_nodes)

nx.draw(core, with_labels=True,node_color='#7209b7',
        node_size=1600)

```

```python
D = nx.DiGraph()
D.add_edges_from([
    (1,2),
    (2,3),
    (3,2), (3,4), (3,5),
    (4,2), (4,5), (4,6),
    (5,6),
    (6,4),
])
nx.draw(D, with_labels=True,node_color='#7209b7',
        node_size=1600)

```

```python
nx.has_path(D, 1, 4)
```

```python
nx.has_path(D, 4, 1)
```

```python
nx.shortest_path(D, 2, 5)
```

```python
nx.shortest_path(D, 5, 2)
```

```python
nx.is_strongly_connected(D)
```

```python
nx.is_weakly_connected(D)
```

```python
list(nx.weakly_connected_components(D))
```

```python
list(nx.strongly_connected_components(D))
```
